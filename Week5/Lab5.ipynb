```python
# Q1) Circle Area and Circumference :
import math
def circle_area (radius) :                                  #  Uses a function to calculate the area.
    return math.pi *  radius ** 2                           # Uses the math module for œÄ and rounding.
circle_circumference = lambda r : 2 * math.pi * r           # Uses a lambda function to calculate the circumference.
radius = float (input("Enter the radius of the circle :"))  # Asks the user to enter the radius of a circle.

area = circle_area(radius)
circumference = circle_circumference(radius)

print(f"Area: {round(area, 2)}")
print(f"Circumference: {round(circumference, 2)}")          # Prints the area and circumference with 2 decimal places.
```

```python
# Q2) Sum and Double Numbers in a String :
import re
def sum_numbers(numbers):                                                 #  Uses a function to sum all numbers found.
    return sum(numbers)
double_sum = lambda x: x * 2                                              #  Uses a lambda function to multiply the sum by 2.
user_string = input("Enter a string: ")                                   # Asks the user to enter a string.
numbers_found = [int(num) for num in re.findall(r'\d+', user_string)]     # Uses regular expressions to find all numbers in the string.

total = sum_numbers(numbers_found)
doubled = double_sum(total)
                                                                          # Prints :
print("Numbers found:", numbers_found)                                    # The numbers found 
print("Sum:", total)                                                      # The sum
print("Doubled value:", doubled)                                          # The doubled value
```

```python
# Q3) Squares and Math Calculations :
import math
def square_numbers(numbers):                                   #  Uses a function to calculate the square of each number using lambda
    return list(map(lambda x: x**2, numbers))
    
nums_input = input("Enter numbers separated by commas: ")      #  Asks the user to enter a list of numbers separated by commas.
numbers = [float(x.strip()) for x in nums_input.split(",")]    # Converts the input into a list of floats.

squares = square_numbers(numbers)                              # Uses math module functions to calculate:
max_square = max(squares)                                      # 1-The maximum square.
sqrt_sum_squares = math.sqrt(sum(squares))                     # 2-The square root of the sum of squares.

print("Squares:", squares)                                     # Prints all results.
print("Maximum square:", max_square)
print("Square root of sum of squares:", round(sqrt_sum_squares, 2))
```

```python
# Q4) Password Strength and Average Score :
import re
import math
def is_strong_password(pw):                                      # Uses regex to check if the password contains:
    if (re.search(r'[A-Z]', pw) and                              # At least one uppercase letter.
        re.search(r'[a-z]', pw) and                              # At least one lowercase letter
        re.search(r'\d', pw)):                                   # At least one number
        return True
    return False
    
def average(scores):
    return sum(scores) / len(scores)
avg_lambda = lambda scores: average(scores)                      # Uses a function and lambda to calculate the average score.

password = input("Enter your password: ")
strength = "Strong" if is_strong_password(password) else "Weak"  

scores_input = input("Enter 5 scores separated by spaces: ")     # Asks the user to enter 5 scores separated by spaces.
scores = [float(x) for x in scores_input.split()]
avg_score = math.ceil(avg_lambda(scores))                        # Uses math.ceil to round the average to the nearest integer.

print("Password strength:", strength)                            # Prints whether the password is strong or weak.
print("Rounded average score:", avg_score)                       # Print the rounded average score.
```

```python
# Q5) Vowel Words and Average Length :
import re
import math
def count_vowels(text):                                                   # Uses a function to count total vowels in the paragraph.
    return len(re.findall(r'[aeiouAEIOU]', text))
word_length = lambda word: len(word)                                      # Uses lambda to calculate the length of each vowel word.
paragraph = input("Enter a paragraph: ")                                  # Asks the user to enter a paragraph of text.
vowel_words = re.findall(r'\b[aeiouAEIOU]\w*', paragraph)                 # Uses regex to find all words starting with a vowel.
total_vowels = count_vowels(paragraph)
lengths = [word_length(word) for word in vowel_words]
average_length = round(sum(lengths) / len(lengths), 2) if lengths else 0  # Uses math functions to calculate the average word length (round to 2 decimals).
                                                                          # Prints:
print("Words starting with vowels:", vowel_words)                         # Words starting with vowels.
print("Total vowels:", total_vowels)                                      # Total vowels.
print("Average length of vowel-words:", average_length)                   # Average vowel-word length
```
