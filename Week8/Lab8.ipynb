```python
# PART 8-1
# 1. Create and inspect array A
import numpy as np

A = np.arange(1, 21).reshape(4, 5)

print("A:\n", A)
print("dtype:", A.dtype)
print("ndim:", A.ndim)
print("shape:", A.shape)
print("size:", A.size)
print("itemsize:", A.itemsize)

B = A.reshape(5, 4)
print("\nB:\n", B)
```

```python
#2. Broadcasting and conditions :
a = np.array([[0.,0.,0.],
              [10.,10.,10.],
              [20.,20.,20.],
              [30.,30.,30.]])

b = np.array([0., 1., 2.])

c = a + b
print("c:\n", c)

mask = (c >= 22)
print("\nmask:\n", mask)

count = np.sum(mask)
print("\nCount of values >= 22:", count)
```

```python
#3. Grades statistics :
grades = np.array([[87, 96, 70],
                   [100, 87, 90],
                   [94, 77, 90],
                   [100, 81, 82]])

print("Sum:", np.sum(grades))
print("Min:", np.min(grades))
print("Max:", np.max(grades))
print("Mean:", np.mean(grades))
print("Std:", np.std(grades))
print("Var:", np.var(grades))

print("\nMean per column:", np.mean(grades, axis=0))
print("Mean per row:", np.mean(grades, axis=1))
print("Max per row:", np.max(grades, axis=1))
```

```python
#4. Custom 3×3 array :
arr = np.array([[2, 4, 6],
                [1, 3, 5],
                [7, 9, 11]])

print("Row sums:", np.sum(arr, axis=1))
print("Column sums:", np.sum(arr, axis=0))
print("Overall mean:", np.mean(arr))
```

```python
#5. Array operations :
A = np.array([[10, 20, 30],
              [40, 50, 60]])

B = np.array([1, 2, 3])

print("A + B:\n", A + B)
print("\nA * 2:\n", A * 2)
print("\nA - 5:\n", A - 5)
```

```python
#6. Index operations :
grades = np.array([[70, 85, 90],
                   [88, 76, 95],
                   [60, 80, 72]])

# Index of largest value in whole array
print("Index of max (flattened):", np.argmax(grades))

# Convert to 2D index
print("2D index of max:", np.unravel_index(np.argmax(grades), grades.shape))

# Highest grade in each row
print("Index of max per row:", np.argmax(grades, axis=1))

# Lowest grade in each column
print("Index of min per column:", np.argmin(grades, axis=0))
```

```python
# 4. Random matrix + unique values :
# A) Random matrix :
np.random.seed(0)

matrix = np.random.randint(-5, 6, (4, 4))
print(matrix)
```

```python
# 5. Indexing & reshaping :
x = np.array([2, 0, 1, 5, 4, 1, 9])

A = np.array([
    [3, 8, 1],
    [6, 2, 9]
])
```
