```python
# Q1) Sum of Even Positive Number :
total = 0
for i in range (10) :
    num = int(input("Enter a number :")) # Asks the user to enter 10 numbers.
    if num == 0 :
        break                            # Stops taking input if the user enters 0 using break.
    if num < 0 :
        continue                         # Skips negative numbers using continue.
    if num % 2 == 0 :
        total += num                     # Adds only even positive numbers using an augmented assignment operator.
print ("Final sum :", total)             # Prints the final sum.
```

```python
# Q2) Multiplication Table :
num = int (input("Enter a number :"))    # Asks the user to enter a number.
for i in range (1,13):                   # Prints its multiplication table from 1 to 12 using a for loop.
    product = num * i
    if i % 3 == 0 :
        continue                         # Skips printing multiples of 3 using continue.    
    if product > 50 :
        break                            # Stops the table if the product is greater than 50 using break.   
print (f"{num} x {i} = {product}")        
```

```python
# Q3) Guess the Secret Number :
secret = 15                                                   # Chooses a secret number between 1 and 20.
attempts = 0
while True :                                                  # Uses a while loop to allow unlimited attempts. 
    guess = int (input("Guess a number between 1 and 20 :"))  # Asks the user to guess the number.
    if guess < 1 or guess > 20 :
        continue                                              # Skips invalid guesses (less than 1 or greater than 20) using continue. 
    attempts += 1                                             # Keeps track of the number of valid attempts using an augmented assignment (+=).
    if guess == secret :
        print ("Correct !ğŸ‰")
        break                                                 # Stops the loop if the user guesses correctly (break). 
    else :
        print ("Try again ğŸ˜£")
print ("Number of valid attempts :",attempts)        
```

```python
# Q4) Count Number Divisible by 4 :
n = int(input("Enter a number :"))   # Asks the user to enter a number n.
count = 0
for i in range (1, n + 1):
    if i > 50 :
        break                        # Stops counting if a number greater than 50 is found (break). 
    if i % 8 == 0 :
        continue                     # Skips numbers divisible by 8 using continue.
    if i % 4 == 0 :                  # Counts how many numbers from 1 to n are divisible by 4
        count += 1
print ("Count :", count)             # Prints the count at the end. 
```
