```python
# Q1) List operations: sum, average, max, min :
numbers = []
for i in range(5):
    num = float(input(f"Enter number {i+1}: "))    # Asks the user to enter 5 numbers and stores them in a list.
    numbers.append(num)
                                                   # Calculates :  
total = sum(numbers)                               # The sum ,                             
average = total / len(numbers)                     # Average ,
maximum = max(numbers)                             # Maximum ,
minimum = min(numbers)                             # Minimum using list operations. 

print("Numbers:", numbers) 
print("Sum:", total)
print("Average:", round(average, 2))               # Prints the results.
print("Maximum:", maximum)
print("Minimum:", minimum)
```

```python
# Q2) Tuple operations :
words = tuple(input(f"Enter word {i+1}: ") for i in range(5))   # Asks the user to enter 5 words and stores them in a tuple.
                                                                # Prints : 
print("First word:", words[0])                                  # The first word.                   
print("Last word:", words[-1])                                  # The last word.
print("Total words:", len(words))                               # The total number of words.

check_word = input("Enter a word to check: ")                   # Checks if a specific word (entered by the user) is in the tuple.
if check_word in words:
    print(check_word, "is in the tuple")
else:
    print(check_word, "is not in the tuple")
```

```python
# Q3) Dictionary of student scores :
students = {}                                         # Creates a dictionary to store student names and their scores.
for i in range(3):
    name = input(f"Enter student {i+1} name: ")       # Asks the user to enter 3 students and their scores
    score = float(input(f"Enter {name}'s score: "))
    students[name] = score
print("All students and scores:", students)           # Prints all students and their scores.

highest_student = max(students, key=students.get)     # Finds and prints the student with the highest score.
print("Highest score student:", highest_student, "with score", students[highest_student])
```

```python
# Q4) Set operations :
list1 = list(map(int, input("Enter numbers for first list separated by space: ").split()))  # Asks the user to enter two lists of numbers.
list2 = list(map(int, input("Enter numbers for second list separated by space: ").split()))

set1 = set(list1)     # Converts them to sets.
set2 = set(list2)
                                                                                            # Prints : 
print("Union:", set1 | set2)                                                                # The union ,        
print("Intersection:", set1 & set2)                                                         # Intersection ,
print("Difference (set1 - set2):", set1 - set2)                                             # Difference of the two sets.
print("Difference (set2 - set1):", set2 - set1)
```

```python
# Q5) Names list, tuple, dictionary, set :
names_list = [input(f"Enter name {i+1}: ") for i in range(5)]   # Asks the user to enter 5 names and stores them in a list.

names_tuple = tuple(names_list)                                 # Converts the list to a tuple and prints it.
print("Names tuple:", names_tuple)

names_dict = {name: len(name) for name in names_list}           # Creates a dictionary with each name as a key and their name length as the value.
print("Names dictionary with lengths:", names_dict)

unique_lengths = set(names_dict.values())                       # Creates a set of unique name lengths and prints it.
print("Unique name lengths:", unique_lengths)
```
