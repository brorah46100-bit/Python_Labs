```python
# Question 2: Data Types and Variables:
```

```python
# Text (str) :
student_name = "Abrar"
print(student_name)
print(type(student_name))
```

```python
# Numeric (int) :
age = 20
print(age, type(age))
```

```python
# Numeric (float):
gpa = 3.75
print(gpa, type(gpa))
```

```python
# Sequence (list) :
subjects = ["Math", "Python",  "Physics"]
print(subjects)
print(type(subjects))
```

```python
# Mapping (dict) :
student_info = { "id":101 ,"year": 2}
print(student_info)
print(type(student_info))
```

```python
# Set :
skills = {"Python","SQL"}
print(skills)
print(type(skills))
```

```python
# Boolean :
is_enrolled = True
print(is_enrolled)
print(type(is_enrolled))
```

```python
# None Type :
graduation_data= None
print(graduation_data)
print(type(graduation_data))
```
