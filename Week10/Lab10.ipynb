```python
# Part 1: Line Plot Basics :
# Import necessary libraries :
import matplotlib.pyplot as plt
import numpy as np
```

```python
x1 = [1, 2, 3, 4, 5]
y1 = [2, 4, 6, 8, 10]

plt.figure(figsize=(6,4))
plt.plot(x1, y1, marker='o', linestyle='-')  # Line plot with markers
plt.title("Line Plot Example")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.grid(True)
plt.show()
```

```python
# Part 2: Scatter Plot and Customization :
x2 = [1, 2, 3, 4, 5]
y2 = [5, 7, 6, 8, 10]

plt.figure(figsize=(6,4))
plt.scatter(x2, y2, s=100, c='red', alpha=0.7)  # size=100, color=red
plt.title("Scatter Plot Example")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.grid(True)
plt.show()
```

```python
# Part 3: Multiple Lines and Legend :
x3 = np.array([1, 2, 3, 4, 5])
y_linear = x3
y_quadratic = x3**2

plt.figure(figsize=(6,4))
plt.plot(x3, y_linear, marker='o', linestyle='-', label='Linear')
plt.plot(x3, y_quadratic, marker='s', linestyle='--', label='Quadratic')
plt.title("Multiple Lines with Legend")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Part 4: Bar Chart and Pie Chart :
# Bar Chart :
categories = ['A', 'B', 'C']
values_bar = [10, 20, 15]

plt.figure(figsize=(6,4))
plt.bar(categories, values_bar, color=['blue', 'green', 'orange'])
plt.title("Bar Chart Example")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.show()
```

```python
# Pie Chart :
values_pie = [40, 35, 25]
labels_pie = ['Python', 'Java', 'C++']

plt.figure(figsize=(6,6))
plt.pie(values_pie, labels=labels_pie, autopct='%1.1f%%', startangle=90, colors=['gold', 'lightblue', 'lightgreen'])
plt.title("Pie Chart Example")
plt.show()
```

```python
# Part 5: Subplots and Layout :
x5 = np.array([1, 2, 3, 4, 5])
y5_1 = x5
y5_2 = x5**2

plt.figure(figsize=(10,4))
plt.subplot(1, 2, 1)  # 1 row, 2 columns, subplot 1
plt.plot(x5, y5_1, marker='o', color='purple')
plt.title("y = x")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.grid(True)

plt.subplot(1, 2, 2)  # subplot 2
plt.plot(x5, y5_2, marker='s', color='teal')
plt.title("y = x²")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.grid(True)

plt.tight_layout()  # prevent overlap
plt.show()
```

```python
# Part 6: Histogram, Box Plot, and Save Figure :
# Generate random data :
data = np.random.randn(100)  
```

```python
# Histogram
plt.figure(figsize=(6,4))
plt.hist(data, bins=10, color='skyblue', edgecolor='black')
plt.title("Histogram of Random Data")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()
```

```python
# Box Plot
plt.figure(figsize=(4,6))
plt.boxplot(data, patch_artist=True, boxprops=dict(facecolor='lightgreen'))
plt.title("Box Plot of Random Data")
plt.ylabel("Value")
plt.grid(True)
plt.show()
```

```python
# Save one of the figures
plt.figure(figsize=(6,4))
plt.hist(data, bins=10, color='pink', edgecolor='black')
plt.title("Histogram Saved to File")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.grid(True)
plt.savefig("histogram.png") 
plt.close()
```
