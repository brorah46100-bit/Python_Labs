```python
### Q5) Calculator menu using match :
print ("Menu :")
print ("1. Addition â•")
print ("2. Subtraction â–")
print ("3. Multiplication âœ–ï¸")
print ("4. Division â—")

Choice = input ("Choose an option (1-4) :")

num1 = float (input ("Enter first number :"))
num2 = float (input ("Enter second number :"))

match Choice :
    case "1" :
        result = num1 + num2
    case "2" :
        result = num1 - num2
    case "3" :
        result = num1 * num2
    case "4" :
        if num2 !=0 :
            result = num1 / num2
        else :
            result = "Cannot divide by ZERO"
    case _:
        result = "INVALID CHOICE âŒ"
print ("Result :" ,result)        
```
